```python
# Renaming columns in 2016 Table

df_2016.rename(columns = {'Country' : 'country'}, inplace = True)
df_2016.rename(columns = {'Happiness Rank' : 'happiness_rank'}, inplace = True)
df_2016.rename(columns = {'Happiness Score' : 'happiness_score'}, inplace = True)
df_2016.rename(columns = {'Economy (GDP per Capita)' : 'economy'}, inplace = True)
df_2016.rename(columns = {'Family' : 'family'}, inplace = True)
df_2016.rename(columns = {'Health (Life Expectancy)' : 'health'}, inplace = True)
df_2016.rename(columns = {'Freedom' : 'freedom'}, inplace = True)
df_2016.rename(columns = {'Trust (Government Corruption)' : 'trust'}, inplace = True)
df_2016.rename(columns = {'Generosity' : 'generosity'}, inplace = True)
```

```python
# Renaming columns in 2017 Table

df_2017.rename(columns = {'Country' : 'country'}, inplace = True)
df_2017.rename(columns = {'Happiness.Rank' : 'happiness_rank'}, inplace = True)
df_2017.rename(columns = {'Happiness.Score' : 'happiness_score'}, inplace = True)
df_2017.rename(columns = {'Economy..GDP.per.Capita.' : 'economy'}, inplace = True)
df_2017.rename(columns = {'Family' : 'family'}, inplace = True)
df_2017.rename(columns = {'Health..Life.Expectancy.' : 'health'}, inplace = True)
df_2017.rename(columns = {'Freedom' : 'freedom'}, inplace = True)
df_2017.rename(columns = {'Trust..Government.Corruption.' : 'trust'}, inplace = True)
df_2017.rename(columns = {'Generosity' : 'generosity'}, inplace = True)
```
