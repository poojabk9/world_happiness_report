## Q6. Create a categorical plot and interpret the results.

```python
hist = sns.histplot(df['happiness_score'], bins = 10, kde = True)
```

```python
hist.figure.savefig(os.path.join(path, 'Visualizations', 'categorical.png'))
```

```python
# Creating category by Happiness Score

df.loc[df['happiness_score'] < 4, "Country's Happiness"] = 'Low'
df.loc[(df['happiness_score'] >= 4) & (df["happiness_score"] <= 6), "Country's Happiness"] = 'Midium'
df.loc[df['happiness_score'] > 6, "Country's Happiness"] = 'High'
```

```python
df["Country's Happiness"].value_counts(dropna = False)
```

```python
df.head()
```

```python
# Create a categorical plot

sns.set(style="ticks")
g = sns.catplot(x="year", y="happiness_score", hue="Country's Happiness", data=df)
```
