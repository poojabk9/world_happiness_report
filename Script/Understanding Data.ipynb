```python
# Importing libraries

import pandas as pd
import numpy as np
import os
```

```python
# Importing datasets

path = r'/Users/neroshakrishna/Desktop/P/Learning/Data_Analysis/CareerFoundry/Data Immersion/Achievement_6/WorldHappiness'
df = pd.read_csv(os.path.join(path, 'Data', 'Prepared Data', 'merged_data.csv'), index_col = False)
```

```python
# Display the first few rows of the dataset

df.head()
```

```python
# Removing Column 'Unnamed: 0'

df = df.drop(columns = ['Unnamed: 0'])
```

```python
# Summary statistics for numerical columns

df.describe()
```

```python
# Check for mixed-type data

for col in df.columns.tolist():
    weird = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df[weird]) > 0:
        print (col)
```

```python
# Data Profile


data_profile = {
    "Number of rows": len(df),
    "Number of columns": len(df.columns),
    "Column names": list(df.columns),
    "Data types": df.dtypes.to_dict(),
    "Missing values": df.isnull().sum().to_dict(),
    "Unique values": df.nunique().to_dict(),
    "Summary statistics": df.describe().to_dict(),
}
```
